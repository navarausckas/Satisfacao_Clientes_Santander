# Projeto 3 - Prevendo Nível de Satisfação de Clientes Santander

```python
#Importando pacotes que serão utilizados no projeto
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
#Carregando os dados
dados = pd.read_csv('dadoshistoricos.csv')
```

```python
# Avaliando o tipo de objeto
type(dados)
```

```python
#Removendo a coluna ID
dados = dados.drop('ID', axis=1)
dados.shape
```

```python
#Exibindo os dados
dados
```

```python
#Separando a variavel TARGET
TARGET = dados[['TARGET']]
```

```python
#Criando uma lista com as colunas do objeto
colunas = dados.columns[0:369].tolist()
```

```python
#Visualizando a lista com os nomes das colunas
colunas
```

```python
#Criando um array dos dados
array = dados.values
```

```python
#Visualizando o tipo do objeto
type(array)
```

```python
# Separando o array em componentes de input (X) da variavel TARGET
X = array[:,0:369]


# Gerando a nova escala (normalizando os dados)
scaler = MinMaxScaler(feature_range = (0, 1))
rescaledX = scaler.fit_transform(X)
```

```python
#Transformando o Array em Dataframe
dadosNormalizados = pd.DataFrame(rescaledX, columns = colunas)
```

```python
#Visualizando o formato do dataframe
dadosNormalizados.shape
```

```python
#Reduzindo a dimensionalidade
pca = PCA(n_components = 30) 
dados = pca.fit_transform(dadosNormalizados)
```

```python
#Convertendo o objeto PCA para um dataframe
dados = pd.DataFrame(dados)
```

```python
#Concatetando a variavel TARGET ao Dataframe
dados = pd.concat([dados, TARGET], axis=1)
```

```python
#Visualizando os dados
dados
```

```python
# Dados de Treino e de Teste
dados_treino = dados.sample(frac=0.60, random_state=0)
dados_teste = dados.drop(dados_treino.index)
```

```python
#Gerando as variaveis preditoras e variavel target
x_treino = dados_treino.iloc[:,0:30]
y_treino = dados_treino.iloc[:,30:31]

x_teste = dados_teste.iloc[:,0:30]
y_teste = dados_teste.iloc[:,30:31]
```

```python
#Contagem dos dados
dados_treino.count()
```

```python
#Contagem dos dados
dados_teste.count()
```

```python
# Classificador Random Forest
classificador = RandomForestClassifier(max_depth=5, random_state=0)
classificador.fit(x_treino, y_treino)

# Executando as previsões
y_pred = classificador.predict(x_teste)
```

```python
# Avaliando a Confusion Matrix e a Acuracia
cm = confusion_matrix(y_teste, y_pred)
print(cm)
print('Acuracia: %s' % accuracy_score(y_teste, y_pred))
```
